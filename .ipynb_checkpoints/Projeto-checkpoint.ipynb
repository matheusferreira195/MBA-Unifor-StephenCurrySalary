```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('espn-nba-players-data/NBA_Players.csv',sep=',')
```

```python
data.columns = data.columns.str.replace(' ', '')
```

```python
data.dtypes
```

```python
dados_preproc = data.loc[data['SALARY'] != 'Not signed']
```

```python
dados_preproc['SALARY'] = dados_preproc['SALARY'].apply(lambda x: float(x.replace(',','')))
```

```python
dados_preproc['SALARY'] 
```

## Pergunta: o que melhor prevê o salário de um jogador de NBA?
## O salário do jogador x condiz com a previsão?

```python
corrMatrix = dados_preproc.corr()
```

#### Melhores preditores do salário

```python
corrMatrix['SALARY'].loc[corrMatrix['SALARY']>0.5]
```

```python
fig, ax = plt.subplots(figsize=(20,20))
sn.heatmap(corrMatrix, annot=True,ax=ax)
```
